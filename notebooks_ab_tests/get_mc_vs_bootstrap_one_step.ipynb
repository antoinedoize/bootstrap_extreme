## Bootstrap variance vs Monte-Carlo variance for one step : de_hann_98 + kulik 

```python
from main_functions import *


list_methods = [
    {"method_name": "de_hann_98",
     "method": get_bootstrap_variance_de_hann_98,
     "size_sample_bootstrap_ratio": 0.8
    },
    {"method_name": "kulik",
     "method": get_bootstrap_variance_kulik,
     "size_sample_bootstrap_ratio": 1
    }
] 

inv_gamma = 3
nb_monte_carlo_steps = 50
nb_bootstrap_steps = 50
size_sample = 10000
output_dict = get_bootstrap_vs_monte_carlo(nb_monte_carlo_steps,
                                nb_bootstrap_steps,
                                inv_gamma,
                                size_sample,
                                list_methods,
                                k_order_statistic=int(0.1*size_sample))


print("mc_hill_std",output_dict["monte_carlo_std"])
print("mc_hill",output_dict["m_c_hill_estimator_mean"])
print("de_hann_98_std",np.mean(output_dict["de_hann_98_std"]))
print("kulik_std",np.mean(output_dict["kulik_std"]))     

plt.plot(output_dict["de_hann_98_std"], label="de_hann_98")
plt.plot(output_dict["kulik_std"], label="kulik_std")
plt.plot([output_dict["monte_carlo_std"] for i in range(len(output_dict["kulik_std"]))], label="mc_std")
plt.title("Standard deviation for each monte carlo step")
plt.legend()
plt.show()
plt.plot(output_dict["k_order_statistic"], label="k_order_statistic")
plt.title("k_order_statistic for each monte carlo step")
plt.legend()
plt.show()
```
